# Drawing the `Distance Line`

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
start = 1
stop = 101
size = 100

x = np.random.randint(start, stop, size)
x_mean = x.mean()
x_mean
```

### Single Color

```python
fig = plt.figure(figsize= (20, 10))

plt.plot(x, 'o')
plt.plot([0, size], [x_mean, x_mean], '.-.', linewidth= 5, label= 'Mean line')

plt.title("Single color")
plt.xticks(range(0, stop, 5))
plt.xlabel("Indices")
plt.ylabel("Values")

for i, xi in enumerate(x):
    plt.plot([i, i], [xi, x_mean], color= 'red', alpha= 0.5)
    
plt.legend()
```

### Multi Color

```python
fig = plt.figure(figsize= (20, 10))

plt.plot(x, 'o')
plt.plot([0, size], [x_mean, x_mean], '.-.', linewidth= 5, label= 'Mean line')

plt.title("Multi color")
plt.xticks(range(0, stop, 5))
plt.xlabel("Indices")
plt.ylabel("Values")

for i, xi in enumerate(x):
    plt.plot([i, i], [xi, x_mean], alpha= 0.5)
    
plt.legend()
```
