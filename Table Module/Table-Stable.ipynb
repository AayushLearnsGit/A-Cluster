#### I have considered the following points that the table can accomplish:  
• The table should be flexible i.e. N number of columns and N number of rows can be stored.  
• Should work to show almost any kind of grid data.  
• Should and must maintain the column width.  
• Can give the optional header names. ( width considered and affected to that whole column )  
• Easy to read and general purpose.

## Data must be 2D (obviously!)

## Let's talk headers.

```python
# All kind of numerical data
data = np.random.randint(0, 10000, (50, 5))
tbl = Table(data, headers= ['A', 'B', 'C', 'D', 'E'])
tbl.construct()
```

### Headers are given importannnnnccccceeeeee............

```python
# All kind of numerical data
data = np.random.randint(0, 10000, (50, 5))
tbl = Table(data, headers= ['AAAAAAA', 'Bbbbbbb', 'C121212', 'D                       121212125', 'E'])
tbl.construct()
```

# You know what?
You can toggle the rows!  
Use: `tbl.construct(rows= True)`

```python
data = np.random.randint(0, 10000, (10, 5))
tbl = Table(data, headers= ['A', 'B', 'C', 'D', 'E'])
tbl.construct(rows= True)
```

###  Text!

```python
# Text data

data = [[1, 'Aayush', 'Male', 180.6, 'Bhuj'],
        [2, 'Hsuyaa', 'Female', 18.6, 'Pluto'],
        [3, 'Aayushi', 'Female', 160.6, '????'],
        [4, 'Ihsuyaa', 'Elam', 140.6, 'Venus'],
        [5, 'Narendra Modi', 'Male', 180.6, 'Delhi'],]
tbl = Table(data, headers= ['Id', 'Name', 'Gender', 'Height', 'Place'])
tbl.construct()
```
